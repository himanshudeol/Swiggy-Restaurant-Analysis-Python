## Easy Questions
### 1. Count Restaurants:
        - How many restaurants are in the dataset?
### 2. Find Maximum Price:
        - What is the highest price of a restaurant in the dataset?
### 3. Average Ratings:

### 4. Total Ratings:

```python
import pandas as pd

df = pd.read_csv("swiggy.csv")
df.info()
df.head(10)
```

```python
#1. Count Restaurants
restaurant_count = df["Restaurant"].nunique()
print("Number of Restaurants:", restaurant_count)
```

```python
#2. Find Maximum Price
max_price = df["Price"].max()
print("Highest Restaurant Price:", max_price)
```

```python
#3. Average Ratings
average_rating = df["Avg ratings"].mean()
print("Average Restaurant Rating:", average_rating)
```

```python
# 4.Total Ratings
total_ratings = df["Total ratings"].sum()
print("Total Ratings Across All Restaurants:", total_ratings)
```

```python
# 5.Food Type Count
food_type_count = df["Food type"].nunique()
print("Number of Different Food Types:", food_type_count)
```

#### The swiggy dataset consists of following columns: 
    * ID – A unique identifier for each restaurant.
    * Area – The specific locality or neighborhood where the restaurant is located.
    * City – The city where the restaurant operates.
    * Restaurant – The name of the restaurant.
    * Price – The average cost for two people at the restaurant (in ₹).
    * Avg ratings – The average customer rating of the restaurant (out of 5).
    * Total ratings – The total number of ratings/reviews the restaurant has received.
    * Food type – The type of cuisine(s) served by the restaurant (e.g., Chinese, Indian, Italian).
    * Address – The specific street address of the restaurant.
    * Delivery time – The estimated delivery time (in minutes) for online food orders from the restaurant.

## Hard Questions
### 1. Correlation Analysis:
        - Is there a correlation between restaurant prices and average ratings? Calculate the correlation coefficient and visualize it with a scatter plot.
### 2. Delivery Time Outliers:
        - Identify any outliers in delivery time using a box plot and explain what these outliers might indicate about those restaurants.
### 3. Price and Ratings Box Plot:
        - Create a box plot comparing the distribution of prices for restaurants with different average rating categories (e.g., below 3, 3-4, above 4).
### 4. Grouped Analysis:
        - Perform a grouped analysis to determine the average price and average rating for each city and visualize the results with a grouped bar chart.

```python
#1. Correlation Analysis: Price vs. Average Ratings
correlation = df["Price"].corr(df["Avg ratings"])
print(f"Correlation Between Price and Ratings: {correlation:.2f}")

#Scatter plot for visualization
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Price"], y=df["Avg ratings"], alpha=0.5)
plt.xlabel("Restaurant Price (₹)")
plt.ylabel("Average Rating")
plt.title(f"Correlation between Price and Ratings (r = {correlation:.2f})")
plt.show()
```

```python
#2. Delivery Time Outliers: Box plot
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Delivery time"], color="red")
plt.xlabel("Delivery Time (minutes)")
plt.title("Box Plot of Delivery Time")
plt.show()
```

```python
#3. Price and Ratings Box Plot: Categorizing ratings
df["Rating Category"] = pd.cut(df["Avg ratings"], bins=[0, 3, 4, 5], labels=["Below 3", "3-4", "Above 4"])

plt.figure(figsize=(8, 5))
sns.boxplot(x="Rating Category", y="Price", data=df, palette="coolwarm")
plt.xlabel("Rating Category")
plt.ylabel("Price (₹)")
plt.title("Distribution of Prices for Different Rating Categories")
plt.show()
```

```python
#4. Grouped Analysis: Average price and rating per city
city_grouped = df.groupby("City").agg({"Price": "mean", "Avg ratings": "mean"}).sort_values("Price", ascending=False)

#Plot Grouped Bar Chart
fig, ax1 = plt.subplots(figsize=(12, 6))
ax2 = ax1.twinx()
city_grouped.head(10).Price.plot(kind="bar", color="blue", ax=ax1, position=1, width=0.4)
city_grouped.head(10)["Avg ratings"].plot(kind="bar", color="orange", ax=ax2, position=0, width=0.4)

ax1.set_ylabel("Average Price (₹)")
ax2.set_ylabel("Average Rating")
ax1.set_xlabel("City")
ax1.set_title("Average Price and Rating per City")
plt.show()

print("Top 10 Cities by Average Price and Rating:\n", city_grouped.head(10))
```

#### Insights:
1. Price vs. Ratings Correlation:
    - Weak positive correlation (r = 0.11), meaning higher prices don't strongly indicate better ratings.
2. Delivery Time Outliers:
    - Some restaurants have extremely high delivery times, possibly due to long distances or operational delays.
3. Price vs. Ratings Box Plot:
    - Higher-rated restaurants tend to have slightly higher prices, but pricing varies widely across all rating categories.
4. City-Wise Analysis:
    - Mumbai has the highest average restaurant price, while Bangalore has the highest average rating. Surat has the lowest prices.
